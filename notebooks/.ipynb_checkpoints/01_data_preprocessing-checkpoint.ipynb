```python
# =============================================
# Combine Phishing & Legitimate Email Datasets
# Handles varied column names safely
# =============================================

import pandas as pd
import os

#  Set your data folder path
base_path = '../data/archive/'
files = ['phishing_email.csv', 'SpamAssasin.csv']  # Add more as needed

combined_data = []

for file in files:
    path = os.path.join(base_path, file)
    print(f"\n Loading: {file}")

    try:
        df = pd.read_csv(path, encoding='latin1')
    except Exception as e:
        print(f" Error reading {file}: {e}")
        continue

    print(" Columns found:", df.columns.tolist())

    #  Identify the column that holds the email content
    possible_cols = ['text', 'content', 'message', 'email', 'body']
    text_col = None

    for col in df.columns:
        if col.lower() in possible_cols:
            text_col = col
            break

    if not text_col:
        # Fallback: use first column if nothing matches
        text_col = df.columns[0]
        print(f" Using fallback column: {text_col}")

    #  Rename to standard 'email_text' column
    df = df.rename(columns={text_col: 'email_text'})

    #  Add label
    if 'phishing' in file.lower():
        df['label'] = 1
    elif 'spamassasin' in file.lower():
        df['label'] = 0
    else:
        print(f" Unknown label for {file}")
        continue

    #  Keep only necessary columns
    df = df[['email_text', 'label']]
    combined_data.append(df)

#  Combine all data
if combined_data:
    final_df = pd.concat(combined_data, ignore_index=True)
    print("\n Combined dataset shape:", final_df.shape)
    print(final_df['label'].value_counts())

    #  Save to CSV
    final_df.to_csv('../data/combined_emails.csv', index=False)
    print(" Saved: ../data/combined_emails.csv")

else:
    print(" No valid data to combine.")

```

```python
# 01_data_preprocessing.ipynb
# Email Text Cleaning for Phishing Detection

import pandas as pd
import numpy as np
import re
import nltk
import matplotlib.pyplot as plt
import seaborn as sns

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Step 1: Load Combined Dataset
df = pd.read_csv('../data/combined_emails.csv')
print("Dataset Loaded. Shape:", df.shape)
print(df.head())

# Step 2: Define Cleaning Function
stop_words = set(stopwords.words('english'))

def clean_email(text):
    if pd.isnull(text):
        return ''
    
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words]
    return ' '.join(tokens)

# Step 3: Apply Cleaning
print("\nCleaning text...")
df['clean_text'] = df['email_text'].apply(clean_email)

# Step 4: Save Cleaned Data
df.to_csv('../data/cleaned_emails.csv', index=False)
print("Saved cleaned data to: ../data/cleaned_emails.csv")

# Step 5: Visualize Class Distribution
sns.countplot(x='label', data=df)
plt.title("Distribution: Phishing (1) vs Legitimate (0)")
plt.xlabel("Label")
plt.ylabel("Count")
plt.show()

# Step 6: Preview Cleaned Text
print("\nSample cleaned text:")
print(df[['email_text', 'clean_text']].head())
```
